```python
import pandas as pd
import numpy as np
```

```python
label_path = '/mnt/data/Birdman/samples/labels/noise.csv'
```

```python
start_end_labels = pd.read_csv(label_path)
```

```python
start_end_labels
```

```python
split = []

for index, row in start_end_labels.iterrows():
    split_len = 1.0
    filename = row['File Name']
    timestart = row['Time Start']
    timeend = row['Time End']
    timestart_sliced = np.arange(timestart, timeend, split_len) 
    timeend_sliced = np.arange(timestart + split_len, timeend + split_len, split_len)
    df = pd.DataFrame(data={'start': timestart_sliced, 'end': timeend_sliced})
    df['sound.files'] = filename
    split.append(df)
```

```python
df = pd.concat(split, ignore_index=True)
df = df.rename(columns={'index': 'selec', 'Time Start': 'start', 'Time End': 'end'}).reset_index()
```

```python
df.to_csv('noiseR.csv', index=False)
```
